```python
# Download All data
import wget
from selenium import webdriver
import os
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.common.keys import Keys
import json
from selenium.webdriver.common.alert import Alert
import time
import urllib.request

CHAR_LIST = ['Argenti', 'Arlan', 'Asta', 'Bailu', 'Blade', 'Bronya', 'Clara', 'Dan Heng', 'Dan Heng ? Imbibitor Lunae', 'Dr. Ratio', 'Fu Xuan', 'Gepard', 'Guinaifen', 'Hanya', 'Herta', 'Himeko', 'Hook', 'Huohuo', 'Jing Yuan', 'Jingliu', 'Kafka', 'Luka', 'Luocha', 'Lynx', 'March 7th', 'Natasha', 'Pela', 'Qingque', 'Ruan Mei', 'Sampo', 'Seele', 'Serval', 'Silver Wolf', 'Sushang', 'Tingyun', 'Topaz and Numby', 'Trailblazer', 'Welt', 'Xueyi', 'Yanqing', 'Yukong']

for i in CHAR_LIST:
    # CHANGE SETTINGS AND RUNNING THIS CODE
    # Settings (설정)
    CHARACTOR_TARGET_ID = f'{i}'
    CHARACTOR_LANGUAGE_ID = 'Chinese'
    TARGET_BROWSER = 'Chrome'
    TARGET_SAVE = 'C:\\Users\\starl\\Documents\\tts\\cme'
    TARGET_TYPE_ONLY_TRAVELER = 'Stelle'
    # if you want Aether only, just do it
    # Or Lumine, Chage 'Aether' to 'Lumine'
    # Warning : This Notebook Can't Work on Google Colab (구글 코랩에서는 작동이 허용되지 않을 수 있습니다)

    print(f"now downloading Chractor : {CHARACTOR_TARGET_ID}\nLanguage_ID : {CHARACTOR_LANGUAGE_ID}")
    print(f'Downloading With {TARGET_BROWSER}')

    LINK_URL = f'https://honkai-star-rail.fandom.com/wiki/{CHARACTOR_TARGET_ID}/Voice-Overs/{CHARACTOR_LANGUAGE_ID}'

    option = Options()
    option.add_argument("disable-infobars")
    option.add_argument("disable-extensions")
    user_agent = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_8_2) AppleWebKit/537.17 (KHTML, like Gecko) Chrome/24.0.1309.0 Safari/537.17 "
    option.add_argument('user-agent=' + user_agent)
    option.add_argument('disable-gpu')
    option.add_argument('incognito') 
    option.add_argument('headless')

    s = Service(ChromeDriverManager().install())
    browser = webdriver.Chrome(service=s, options=option)

    browser.get(LINK_URL)
    os.chdir(TARGET_SAVE)
    data_code = []

    if not os.path.exists(TARGET_SAVE + '/' + CHARACTOR_TARGET_ID):
        os.makedirs(TARGET_SAVE + '/' + CHARACTOR_TARGET_ID)

    data = browser.find_elements(By.TAG_NAME, 'table')
    for i in data:
        for j in i.find_elements(By.TAG_NAME, 'td'):
            try:
                text = j.find_element(By.CSS_SELECTOR, 'span:nth-child(3)').text
            except:
                text = 'None'
            audio = j.find_elements(By.CSS_SELECTOR, 'span.audio-button.custom-theme.hidden > a')
            print(audio)
            try:
                os.chdir(TARGET_SAVE + '/' + CHARACTOR_TARGET_ID)
                print("Audio : ", audio[0].get_attribute('href'))
                # print("Audio Downloading...")
                file_name = audio[0].get_attribute('href').split('.ogg')[0].split('/')[-1] + '.ogg'
                print(file_name)
                # wget.download(audio[0].get_attribute('href'))
                if(TARGET_TYPE_ONLY_TRAVELER == 'Stelle'):
                    # print(audio[0].get_attribute('href').replace('Aether', 'Lumine'))
                    wget.download(audio[0].get_attribute('href').replace('Caelus', 'Stelle'))
                else:
                    wget.download(audio[0].get_attribute('href'))
                data_code.append({'file_name' : file_name, 'text' : text})
            except:
                print("Audio : None")
            print("Text : ", text)

    os.chdir(TARGET_SAVE)
    print(data_code)

    # text_generation
    os.chdir(TARGET_SAVE + '/' + CHARACTOR_TARGET_ID)
    with open('data.txt', 'w', encoding='utf-8') as f:
        for item in data_code:
            if(item['text'] == 'None' or item['file_name'] == 'None'):
                continue
            f.write(f'{item["file_name"]}  ||  {item["text"]}\n')
            
    print('Complete Downloading!')
    browser.quit()
```
